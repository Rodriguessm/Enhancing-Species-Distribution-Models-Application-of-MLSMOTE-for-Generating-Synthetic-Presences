# MLSMOTE
Aplicando a técnica de Sobreamostragem de Minoria Sintética Multirótulo para combater o desbalanceamento dos dados.

```python
### Primeiro, é preciso definir quais são os rótulos mais desbalanceados
  ## (i) Para selecioná-los nós calculamos primeiro a razão de desequilíbrio (irpl) de cada rótulo.
  ## (ii) Depois descobrimos qual é a razão de desequilíbrio médio (mir) do conjunto de rótulos.
  ## (iii) Dessa forma, os rótulos que tiverem um irpl abaixo do mir vão seguir como "tail labels", ou seja, os rótulos que precisam ser contrabalanceados

def get_tail_label(train_set):
    columns = train_set.columns
    n = len(columns)
    irpl = np.zeros(n)
    for column in range(n):
        irpl[column] = train_set[columns[column]].value_counts()[1]
    irpl = max(irpl)/irpl
    mir = np.average(irpl)
    tail_label = []
    for i in range(n):
        if irpl[i] > mir:
            tail_label.append(columns[i])
    return tail_label
```

```python
### Para visualizar os rótulos classificados como mais desbalanceados podemos:
tail_labels = get_tail_label(y_train)
print(tail_labels)
```

```python
### Nessa etapa, vamos procurar no dataframe de treino todas as quadrículas (linhas) que nos nossos rótulos minoritários possuem pelo menos uma presença
  ## E salvar o index das linhas em uma lista

def get_index(train_set):
  tail_labels = get_tail_label(train_set)
  index = set()
  for tail_label in tail_labels:
    sub_index = set(train_set[train_set[tail_label]==1].index)
    index = index.union(sub_index)
  return list(index)

index = get_index(y)
print(index)
```

```python
### Cria dois subconjuntos apartir dos rótulos minoritários e dos index
def get_minority_instace(X, y):
    index = get_index(y)
    tail_labels = get_tail_label(y)  # Separando as colunas minoritárias

    X_sub = X[X.index.isin(index)].reset_index(drop=True)
    y_sub = y[y.index.isin(index)][tail_labels].reset_index(drop=True)  # Filtrando colunas a partir do index

    return X_sub, y_sub
```

```python
### Verificando se os dois dataframes estão ok
X_sub, y_sub = get_minority_instace(X, y)
print(X_sub.shape, y_sub.shape)
```

```python
y_sub.head() #visualizando os dados
```

```python
### para encontrar os 5 vizinhos mais próximos de cada presença
def nearest_neighbour(X):
    nbs=NearestNeighbors(n_neighbors=5,metric='euclidean',algorithm='kd_tree').fit(X)
    euclidean,indices= nbs.kneighbors(X)
    return indices
```

```python
### Obtendo os vizinhos mais próximos do subconjunto minoritário
indices = nearest_neighbour(X_sub.values)

### Exibindo os resultados
print(X_sub.shape, y_sub.shape)
print("Índices dos 5 vizinhos mais próximos para cada amostra minoritária:\n", indices)
```

```python
### Criando amostras sintéticas
  ## O método gera amostras sinéticas das classes minoritárias dos rótulos mais desbalanceados baseadas nas suas respostas anteriores e nas respostas dos demais rótulos.

'''
    ## Charte, F. & Rivera Rivas, Antonio & Del Jesus, María José & Herrera, Francisco. (2015).
    ## MLSMOTE: Approaching imbalanced multilabel learning through synthetic instance generation.
    ## Knowledge-Based Systems. -. 10.1016/j.knosys.2015.07.019.
'''

def MLSMOTE(X, y, n_sample, seed= None):

    if seed is not None: #para controlar a aleatoridade das amostras criadas
        random.seed(seed)
        np.random.seed(seed)

    indices2 = nearest_neighbour(X.values)
    n = len(indices2)

    new_X = np.zeros((n_sample, X.shape[1]))
    target = np.zeros((n_sample, y.shape[1]))


    for i in range(n_sample):
        reference_idx = random.randint(0, n-1)  # Índice da amostra de referência
        neighbour_idx = random.choice(indices2[reference_idx, 1:])  # Pegando um vizinho aleatório

        # Pegando as linhas reais do dataframe
        reference = X.iloc[reference_idx]
        neighbour = X.iloc[neighbour_idx]

        # Pegando os índices de todos os vizinhos para a soma no target
        all_point = indices2[reference_idx]
        nn_df = y.iloc[all_point]  # Pegando as labels dos vizinhos

        # Criando novo target com base na soma das labels dos vizinhos
        ser = nn_df.sum(axis=0, skipna=True)
        target[i] = np.array([1 if val > 2 else 0 for val in ser])  # Threshold arbitrário

        # Criando um novo ponto interpolado
        ratio = random.random()
        new_X[i] = np.array(reference + ratio * (neighbour - reference))

    # Convertendo de volta para DataFrame
    new_X = pd.DataFrame(new_X, columns=X.columns)
    target = pd.DataFrame(target, columns=y.columns)

    # Concatenando os dados gerados aos originais
    new_X = pd.concat([X, new_X], axis=0).reset_index(drop=True)
    target = pd.concat([y, target], axis=0).reset_index(drop=True)

    return new_X, target

# Chamando as funções corretamente:
X_sub, y_sub = get_minority_instace(X, y)
X_aug, y_aug = MLSMOTE(X_sub, y_sub, n_sample=100, seed= 42)  # número de amostras sintéticas geradas

print(X_aug.shape, y_aug.shape)  # Verificando os tamanhos e os formatos dos dataframes finais
```

```python
### Unindo os DataFrames
X_final = pd.concat([X, X_aug], axis=0).reset_index(drop=True)
y_final = pd.concat([y, y_aug], axis=0).reset_index(drop=True).fillna(0)

### Exibindo os tamanhos finais
print("Tamanho original:", X.shape, y.shape)
print("Tamanho aumentado:", X_aug.shape, y_aug.shape)
print("Tamanho final:", X_final.shape, y_final.shape)
```

```python
num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ]) # preencher os valores faltante com a média

x_train_modelo = num_pipeline.fit_transform(X_final)
```

```python
x_test_modelo = num_pipeline.transform(x_test)
```

```python
### Transformar os conjuntos respostas em numpy array

y_train_modelo =   np.array(y_final)
y_test_modelo =   np.array(y_test)
```

# DEEP LEARING

```python
### Definindo as métricas
METRICS = [
      keras.metrics.TruePositives(name='tp'),
      keras.metrics.FalsePositives(name='fp'),
      keras.metrics.TrueNegatives(name='tn'),
      keras.metrics.FalseNegatives(name='fn'),
      keras.metrics.BinaryAccuracy(name='accuracy'),
      keras.metrics.Precision(name='precision'),
      keras.metrics.Recall(name='recall'),
      keras.metrics.AUC(name='auc'),
      keras.metrics.AUC(name='prc', curve='PR'), # precision-recall curve
]
```

```python
### Definir a semente para que as futuras replicações possuam os mesmos resultados
seed_value = 42

### Configurar as sementes
os.environ['PYTHONHASHSEED'] = str(seed_value)
random.seed(seed_value)
np.random.seed(seed_value)
tf.random.set_seed(seed_value)
```

```python
### Criar o modelo
model = keras.models.Sequential([
    layers.Dense(128, activation="relu", input_shape=[19], kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed_value)),
    layers.Dropout(0.3, seed=seed_value),
    layers.Dense(64, activation="relu", kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed_value)),
    layers.Dropout(0.3, seed=seed_value),
    layers.Dense(32, activation="relu", kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed_value)),
    layers.Dense(16, activation="relu", kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed_value)),
    layers.Dense(48, activation="sigmoid", bias_initializer='zeros', kernel_initializer=tf.keras.initializers.GlorotUniform(seed=seed_value))
])
```

```python
model.summary()
```

```python
### Nessa etapa nós pedimos para que modelo pare as análises quando 20 análises tiverem métricas com resultados não diferentes significativamente
early_stopping = tf.keras.callbacks.EarlyStopping(patience=20)
```

```python
### Compilando o modelo
model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy", tf.keras.metrics.AUC(name="AUC")])
```

```python
### Fit do modelo
  ## validação de cada análise é feita com 20% do conjunto de treino
  ## o máximo de análises (épocas) que o modelo poderá fazer é 400
  ## batch_size é a quantidade de linhas simultaneas para cada análise
  ## pedimos para que o modelo pare quando as últimas 20 análises não apresentem uma diferença significativa entre elas
baseline_history= model.fit(x_train_modelo, y_train_modelo,
                            validation_split=0.2,
                            epochs=400,
                            batch_size=100,
                            callbacks=early_stopping)
```

```python
### Gráfico com da curva do erro e do AUC

def plot_history(history):
    # Ordenando os valores de erro e AUC
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    auc = history.history['AUC']
    val_auc = history.history['val_AUC']

    epochs = range(1, len(loss) + 1)

    # Gráfico do erro
    plt.figure(figsize=(14, 5))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, loss, label='Loss (train)')
    plt.plot(epochs, val_loss, label='Loss (validation)')
    plt.title('Loss por Época')
    plt.xlabel('Épocas')
    plt.ylabel('Pérdida')
    plt.legend()

    # Gráfico do AUC
    plt.subplot(1, 2, 2)
    plt.plot(epochs, auc, label='AUC (train)')
    plt.plot(epochs, val_auc, label='AUC (validation)')
    plt.title('AUC por Época')
    plt.xlabel('Épocas')
    plt.ylabel('AUC')
    plt.legend()

    plt.tight_layout()
    plt.show()

### Plot dos dois gráficos (gráfico de cotovelo)
plot_history(baseline_history)
```

```python
### Agora com o conjunto de dados teste
predictions = model.predict(x_test_modelo)
```

```python
### Para calcular como cada modelos se saiu por espécie, podemos calcular os AUCs individuais

aucs = {}
species_names = y_test.columns

for i in range(predictions.shape[1]):
    species_name = species_names[i]
    true_labels = y_test_modelo[:, i]
    pred_probs = predictions[:, i]

    if len(np.unique(true_labels)) > 1:
        auc = roc_auc_score(true_labels, pred_probs)
        aucs[species_name] = auc
    else:
        aucs[species_name] = None

### Para mostrar os resultados
for species, score in aucs.items():
    if score is not None:
        print(f"AUC para {species}: {score:.4f}")
    else:
        print(f"AUC para {species}: No se puede calcular AUC (solo una clase presente)")
```
